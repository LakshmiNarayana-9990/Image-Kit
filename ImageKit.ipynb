```python
print(i.imread.__doc__)
x = i.imread('fun.jpg')
```

```python
i.imshow(x)
```

```python
desc = i.get_properties(x)
for name, value in zip(desc.keys(), desc.values()):
    print("%10s  => %1s"%(str(name), str(value)))
```

```python
x1 = i.resize(x, (700, 500))
i.imshow(x1)
```

```python
x2 = i.resize(x, (25,25))
i.get_pixels(x2)
```

```python
r = i.get_band(x, 'r')
i.imshow(r)
g = i.get_band(x, 'g')
i.imshow(g)
b = i.get_band(x, 'b')
i.imshow(b)
```
